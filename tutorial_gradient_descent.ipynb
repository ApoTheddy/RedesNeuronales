```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_funct(x):
  return x**2 - 5*x + 2;

def derivada(x):
  return 2*x - 5;
```

```python
x = np.linspace(-20,20,100)
y = my_funct(x)
```

```python
plt.plot(x,y)
plt.plot(2.5, my_funct(2.5),'bo')
```

```python
init_x = 15
init_y = my_funct(init_x)
plt.plot(x,y)
plt.plot(init_x,init_y,"bo")
```

```python
lr = 0.05
dx = derivada(init_x)
px = init_x - dx * lr
py = my_funct(px)
plt.plot(x,y)
plt.plot(init_x,init_y,"bo")
plt.plot(px,py,"bo") 
```

```python
plt.plot(x,y)
plt.plot(init_x,init_y,"bo")
plt.plot(px,py,"bo")
for i in range(100):
  px = px - derivada(px) * lr
  py = my_funct(px)
  plt.plot(px,py,"bo")  
  print(px)
```
